```python
# Source: https://github.com/jonathanpaulson/AdventOfCode/blob/master/2022/23.py
from collections import defaultdict

# Read a file
f = [x for x in open("puzzle.txt").read().split('\n')]

# Set of all elves' initial position
E = set([(r, c) for r, row in enumerate(f)
                for c, ch in enumerate(row)
                if ch == '#'])

# Define directions
dir_list = ['N', 'S', 'W', 'E']

for t in range(10000):
    any_moved = False

    # P[(r, c)] is the list of elves who want to move to (r, c)
    P = defaultdict(list)

    # For each elf
    for (r, c) in E:

        # If there are no neighbours, don't move
        has_nbr = False
        for dr in [-1, 0, 1]:
            for dc in [-1, 0, 1]:
                if (dr != 0 or dc != 0) and (r + dr, c + dc) in E:
                    has_nbr = True
        if not has_nbr:
            continue

        # Keep track of which elves want to move to a certain coordinate
        moved = False
        for dir in dir_list:
            if (dir == 'N' and (not moved) and
                all(m not in E for m in [(r - 1, c),
                                         (r - 1, c - 1),
                                         (r - 1, c + 1)])):

                P[(r - 1, c)].append((r, c))
                moved = True
            elif (dir == 'S' and (not moved) and
                  all(m not in E for m in [(r - 1, c),
                                           (r - 1, c - 1),
                                           (r - 1, c + 1)])):

                P[(r + 1, c)].append((r, c))
                moved = True
            elif (dir == 'W' and (not moved) and
                all(m not in E for m in [(r - 1, c),
                                         (r - 1, c - 1),
                                         (r - 1, c + 1)])):

                P[(r, c - 1)].append((r,c))
                moved = True
            elif (dir == 'E' and (not moved) and
                  all(m not in E for m in [(r - 1, c),
                                           (r - 1, c - 1),
                                           (r - 1, c + 1)])):

                P[(r, c + 1)].append((r, c))
                moved = True

    # At the end of each round, the first direction becomes
    # the last direction for the next round
    dir_list = dir_list[1:] + [dir_list[0]]

    # For every elf that wants to move
    for k, vs in P.items():

        # If there is only one elf that wants to move to that coordinate
        # Update that elf's position
        if len(vs) == 1:
            any_moved = True
            E.discard(vs[0])
            E.add(k)

    # Part 2
    # Find the first round where the elves cannot move
    if not any_moved:
        print(t + 1)
        break

    # Part 1
    # After 10 rounds, find the smallest rectangle that contains all elves
    # Calculate sum of empty tiles
    if t == 9:
        print(sum([1 for r in range(min(r for (r, _) in E),
                                    max(r for (r, _) in E) + 1)
                     for c in range(min(c for (_, c) in E),
                                    max(c for (_, c) in E) + 1)
                     if (r, c) not in E]))
```
