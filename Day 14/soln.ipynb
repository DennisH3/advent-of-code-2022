```python
# Read a file
f = open("puzzle.txt").read().split("\n")

# First draw out our grid with sands and air otherwise
# To do this, we need grid size. 
# Number of rows = max y-coordinate
# Number of cols = max x-coordinate

# Get the number of columns and rows by splitting the coordinates by "->""
# then flattening the list of lists and stripping white spaces to further 
# split into x- and y-coordinates individually
ncols = max([int(x.split(",")[0]) for x in 
            [item.strip() for sublist in [line.split("->") 
            for line in f] for item in sublist]])

nrows = max([int(x.split(",")[1]) for x in 
            [item.strip() for sublist in [line.split("->") 
            for line in f] for item in sublist]])

# Initialize grid to be just air
G = [['.' for _ in range(ncols+1)] for _ in range(nrows+1)]
print(len(G), len(G[0]))

# Update G to have rocks
# For each path
for sand in [[item.strip() for item in path] for path in [line.split("->") for line in f]]:

    # For the length of each path
    for xy in range(len(sand)-1):
        
        # Get the current coordinates of the path
        cy, cx = sand[xy].split(',')
        cy, cx = int(cy), int(cx)

        # Get the next coordinates of the path
        ny, nx = sand[xy+1].split(',')
        ny, nx = int(ny), int(nx)

        # Check which direction the path goes
        # If next y >= current y, add rocks to the right
        if (ny >= cy):
            for y in range(cy, ny+1):
                G[cx][y] = '#'

        # Else if next y <= current y, add rocks to the left
        elif (ny <= cy):
            for y in range(ny, cy+1):
                G[cx][y] = '#'

        # If next x >= current x, add rocks downwards
        if (nx >= cx):
            for x in range(cx, nx+1):
                G[x][cy] = '#'
        
        # Else if next x <= current y, add rocks upwards
        elif (nx <= cx):
            for x in range(nx, cx+1):
                G[x][cy] = '#'

# Convert 2D array into a set of tuples for each '#'
# R is a set that keeps track of occupied spaces
R = set([(c, r) for c in range(len(G[0])) for r in range(len(G)) if G[r][c] == '#'])

# Add 2 rows
floor = 2+nrows

# Add the floor
lo_x = -1000
hi_x = ncols+1000
for x in range(lo_x, hi_x):
    R.add((x,floor))

# Flag to check if part 1 has been completed
p1 = False

# Drop a lot of sand
for t in range(1000000):

    # Source of sand
    sand = (500,0)
    while True:

        # If sand reaches the floor
        if sand[1]+1>=floor and (not p1):

            # Completed part 1, print amount of sand settled
            p1 = True
            print(t)

        # If space directly down is not occupied
        if (sand[0],sand[1]+1) not in R:

            # Update sand position (Go down)
            sand = (sand[0],sand[1]+1)
        
        # Else if space below and to the left is not occupied
        elif (sand[0]-1,sand[1]+1) not in R:

            # Update sand position (Go diagonally left)
            sand = (sand[0]-1, sand[1]+1)
        
        # Else if space below and to the right is not occupied
        elif (sand[0]+1, sand[1]+1) not in R:

            # Update sand position (Go diagonally right)
            sand = (sand[0]+1, sand[1]+1)
        
        # Else, sand has settled
        else:
            break
    
    # Part 2
    # If sand has reached the source
    if sand == (500,0):

        # Print amount of sand and stop
        print(t+1)
        break

    # Add sand to the set of rocks
    R.add(sand)

# Print output to check
# print('\t'.join([str(i) for i in range(ncols+1)]))
# for i in range(len(G)):
#     print("\t".join(G[i]))
```
